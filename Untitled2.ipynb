```python
import tensorflow as tf

#**************以下函数获取scope_name命名空间下变量名为var_name的变量，不存在创建，存在则返回已存在的变量***********
def get_scope_variable(var_name,shape=None):
  with tf.variable_scope('cur_scope') as scope:            #reuse设置为true不存在会异常，设置为False，存在重名会异常。故我们捕获异常来判断是否存在。
    try:            
      var=tf.get_variable(var_name,shape)
    except ValueError:
      scope.reuse_variables()
      var=tf.get_variable(var_name)
  return var 
var_1 = get_scope_variable("my_var",[100])
var_2 = get_scope_variable("my_var",[100])
print(var_1 is var_2)
print(var_1.name)                                        #此时变量名为  cur_scope/my_var
 
```

```python
import tensorflow as tf
import numpy as np
from numpy import random
x_data=random.random((3,300)).astype('float32')
print(x_data.dtype)
#b1=random.random(45).astype('float32')
def fc_1():
    w1 = tf.get_variable('weight1', shape=[300, 45], initializer=tf.truncated_normal_initializer())
    b1 = tf.get_variable('bias1', shape=[45])
    h_fc_1 = tf.nn.relu(tf.nn.xw_plus_b(x_data, w1, b1), name='relu')
    return h_fc_1
with tf.variable_scope('fc_1') as scope:
    for i in range(3):
        try:
            _output=fc_1()
        except ValueError:
            scope.reuse_variables()
            _output=fc_1()
        print(_output)
        
```

```python

gg=np.array([[0, 0, 0, 0, 0],
 [1, 1, 1, 0, 1],
 [0, 1, 0, 0, 1]])
bb=random.random((3,5,4))
print(bb)
gg=tf.Variable(gg,dtype=tf.float32)
gg=tf.reshape(gg,(3,5,1))
bb=tf.Variable(bb,dtype=tf.float32)
print(gg)
print(bb)
mul=tf.math.multiply(gg,bb)
print(mul)
sess=tf.Session()
sess.run(tf.global_variables_initializer())
print(sess.run(mul))
print(sess.run(tf.reduce_sum(tf.math.multiply(mul,bb),2)))
```

```python
import numpy as np
import tensorflow as tf
label=np.random.randint(0,2,size=(3,15))
print(label)
arr=tf.Variable(label,dtype=tf.int64)
arrlabel=tf.one_hot(arr,depth=2)
sess=tf.Session()
sess.run(tf.global_variables_initializer())
print(sess.run(arrlabel))
pred=np.random.random((3,15,2)).astype('float32')
print(pred)
loss=arrlabel*tf.log(pred)
print(sess.run(loss))
```
